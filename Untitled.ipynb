```python
def plyr_scrape(name, y):
    names = name.split(" ")
    fname = names[0]
    lname = names[1]
    new_name = name.lower()
    sub_name = new_name.replace(" ","-")
    sub_name = sub_name.replace("'","")
    sub_name = sub_name.replace(".","")
    sub_name += "-1.html"
    url = "https://www.sports-reference.com/cbb/players/" + sub_name
    headers = requests.utils.default_headers()
    req = requests.get(url, headers)
    soup = BeautifulSoup(req.content, 'html.parser')
    page = soup.prettify()
    page = soup.find_all("tr")
    page = str(page).split("</tr>")
    x=1
    for th in page:
        if x == 1:
            x += 1
            continue
        elif x == 2:
            x += 1
            name = url[len(root):]
            name = name.replace("-1.html","")
            name = name.split("-")
            fname = name[0].capitalize().strip()
            lname = name[1].capitalize().strip()
            th = str(th).split("data-stat") 
            for i in th:
                if "players_per_game" in i:
                    continue
                elif "/seasons/" in i:
                    season = i.split('.html">')
                    season = season[1][:7]
                    continue
                elif "school_name" in i:
                    school = i.split('.html">')
                    school = school[1]
                    school = str(school).split("<")
                    school = school[0]
                    continue
                elif "conf_abbr" in i:
                    conf = i.split('.html">')
                    conf = conf[1]
                    conf = str(conf).split("<")
                    conf = conf[0]
                    continue
                elif '"g"' in i:
                    g = i.split('g">')
                    g = g[1]
                    g = str(g).split("<")
                    g = g[0]
                    continue
                elif '"gs"' in i:
                    gs = i.split('gs">')
                    gs = gs[1]
                    gs = str(gs).split("<")
                    gs = gs[0]
                    continue
                elif '"mp_per_g"' in i:
                    mp_per_g = i.split('mp_per_g">')
                    mp_per_g = mp_per_g[1]
                    mp_per_g = str(mp_per_g).split("<")
                    mp_per_g = mp_per_g[0]
                    continue
                elif '"fg_per_g"' in i:
                    fg_per_g = i.split('fg_per_g">')
                    fg_per_g = fg_per_g[1]
                    fg_per_g = str(fg_per_g).split("<")
                    fg_per_g = fg_per_g[0]
                    continue
                elif '"fga_per_g"' in i:
                    fga_per_g = i.split('fga_per_g">')
                    fga_per_g = fga_per_g[1]
                    fga_per_g = str(fga_per_g).split("<")
                    fga_per_g = fga_per_g[0]
                    continue
                elif '"fg_pct"' in i:
                    fg_pct = i.split('fg_pct">')
                    fg_pct = fg_pct[1]
                    fg_pct = str(fg_pct).split("<")
                    fg_pct = fg_pct[0]
                    continue
                elif '"fg2_per_g"' in i:
                    fg2_per_g = i.split('fg2_per_g">')
                    fg2_per_g = fg2_per_g[1]
                    fg2_per_g = str(fg2_per_g).split("<")
                    fg2_per_g = fg2_per_g[0]
                    continue
                elif '"fg2a_per_g"' in i:
                    fg2a_per_g = i.split('fg2a_per_g">')
                    fg2a_per_g = fg2a_per_g[1]
                    fg2a_per_g = str(fg2a_per_g).split("<")
                    fg2a_per_g = fg2a_per_g[0]
                    continue
                elif '"fg2_pct"' in i:
                    fg2_pct = i.split('fg2_pct">')
                    fg2_pct = fg2_pct[1]
                    fg2_pct = str(fg2_pct).split("<")
                    fg2_pct = fg2_pct[0]
                    continue
                elif '"fg3_per_g"' in i:
                    fg3_per_g = i.split('fg3_per_g">')
                    fg3_per_g = fg3_per_g[1]
                    fg3_per_g = str(fg3_per_g).split("<")
                    fg3_per_g = fg3_per_g[0]
                    continue
                elif '"fg3a_per_g"' in i:
                    fg3a_per_g = i.split('fg3a_per_g">')
                    fg3a_per_g = fg3a_per_g[1]
                    fg3a_per_g = str(fg3a_per_g).split("<")
                    fg3a_per_g = fg3a_per_g[0]
                    continue
                elif '"fg3_pct"' in i:
                    fg3_pct = i.split('fg3_pct">')
                    fg3_pct = fg3_pct[1]
                    fg3_pct = str(fg3_pct).split("<")
                    fg3_pct = fg3_pct[0]
                    continue
                elif '"ft_per_g"' in i:
                    ft_per_g = i.split('ft_per_g">')
                    ft_per_g = ft_per_g[1]
                    ft_per_g = str(ft_per_g).split("<")
                    ft_per_g = ft_per_g[0]
                    continue
                elif '"fta_per_g"' in i:
                    fta_per_g = i.split('fta_per_g">')
                    fta_per_g = fta_per_g[1]
                    fta_per_g = str(fta_per_g).split("<")
                    fta_per_g = fta_per_g[0]
                    continue
                elif '"ft_pct"' in i:
                    ft_pct = i.split('ft_pct">')
                    ft_pct = ft_pct[1]
                    ft_pct = str(ft_pct).split("<")
                    ft_pct = ft_pct[0]
                    continue
                elif '"orb_per_g"' in i:
                    orb_per_g = i.split('orb_per_g">')
                    orb_per_g = orb_per_g[1]
                    orb_per_g = str(orb_per_g).split("<")
                    orb_per_g = orb_per_g[0]
                    continue
                elif '"drb_per_g"' in i:
                    drb_per_g = i.split('drb_per_g">')
                    drb_per_g = drb_per_g[1]
                    drb_per_g = str(drb_per_g).split("<")
                    drb_per_g = drb_per_g[0]
                    continue
                elif '"trb_per_g"' in i:
                    trb_per_g = i.split('trb_per_g">')
                    trb_per_g = trb_per_g[1]
                    trb_per_g = str(trb_per_g).split("<")
                    trb_per_g = trb_per_g[0]
                    continue
                elif '"ast_per_g"' in i:
                    ast_per_g = i.split('ast_per_g">')
                    ast_per_g = ast_per_g[1]
                    ast_per_g = str(ast_per_g).split("<")
                    ast_per_g = ast_per_g[0]
                    continue
                elif '"stl_per_g"' in i:
                    stl_per_g = i.split('stl_per_g">')
                    stl_per_g = stl_per_g[1]
                    stl_per_g = str(stl_per_g).split("<")
                    stl_per_g = stl_per_g[0]
                    continue
                elif '"blk_per_g"' in i:
                    blk_per_g = i.split('blk_per_g">')
                    blk_per_g = blk_per_g[1]
                    blk_per_g = str(blk_per_g).split("<")
                    blk_per_g = blk_per_g[0]
                    continue
                elif '"tov_per_g"' in i:
                    tov_per_g = i.split('tov_per_g">')
                    tov_per_g = tov_per_g[1]
                    tov_per_g = str(tov_per_g).split("<")
                    tov_per_g = tov_per_g[0]
                    continue
                elif '"pf_per_g"' in i:
                    pf_per_g = i.split('pf_per_g">')
                    pf_per_g = pf_per_g[1]
                    pf_per_g = str(pf_per_g).split("<")
                    pf_per_g = pf_per_g[0]
                    continue
                elif '"pts_per_g"' in i:
                    pts_per_g = i.split('pts_per_g">')
                    pts_per_g = pts_per_g[1]
                    pts_per_g = str(pts_per_g).split("<")
                    pts_per_g = pts_per_g[0]
                    continue
                elif '"sos-dum"' in i:
                    sosdum = i.split('sos-dum">')
                    sosdum = sosdum[1]
                    sosdum = str(sosdum).split("<")
                    sosdum = sosdum[0]
                    continue
                elif '"sos"' in i:
                    sos = i.split('sos">')
                    sos = sos[1]
                    sos = str(sos).split("<")
                    sos = sos[0]
                    sos = sos.strip()
                    _done = "Yes"
                    continue
        elif x==3:
               _done = "No"
    df = pd.DataFrame({"fname":[fname],
                        "lname":[lname],
                        "_done":[_done],
                        "f_yr":[season],
                        "school":[school],
                        "conf":[conf],
                        "g":[g],
                        "gs":[gs],
                        "mp_per_g":[mp_per_g],
                        "fg_per_g":[fg_per_g],
                        "fga_per_g":[fga_per_g],
                        "fg_pct":[fg_pct],
                        "fg2_per_g":[fg2_per_g],
                        "fg2a_per_g":[fg2a_per_g],
                        "fg2_pct":[fg2_pct],
                        "fg3_per_g":[fg3_per_g],
                        "fg3a_per_g":[fg3a_per_g],
                        "fg3_pct":[fg3_pct],
                        "ft_per_g":[ft_per_g],
                        "fta_per_g":[fta_per_g],
                        "ft_pct":[ft_pct], 
                        "orb_per_g":[orb_per_g],
                        "drb_per_g":[drb_per_g],
                        "trb_per_g":[trb_per_g],
                        "ast_per_g":[ast_per_g],
                        "stl_per_g":[stl_per_g],
                        "blk_per_g":[blk_per_g],
                        "tov_per_g":[tov_per_g],
                        "pf_per_g":[pf_per_g],
                        "pts_per_g":[pts_per_g],
                        "sos-dum":[sosdum],
                        "sos":[sos],
                         },index = [y])
    return df
    
```
