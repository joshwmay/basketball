```python
def heights(name, y):
    names = name.split(" ")
    fname = names[0]
    lname = names[1]
    new_name = name.lower()
    sub_name = new_name.replace(" ","-")
    sub_name = sub_name.replace("'","")
    sub_name = sub_name.replace(".","")
    sub_name += "-1.html"
    url = "https://www.sports-reference.com/cbb/players/" + sub_name
    headers = requests.utils.default_headers()
    req = requests.get(url, headers)
    soup = BeautifulSoup(req.content, 'html.parser')
    page = soup.prettify()
    page = soup.find_all("p")
    page = str(page).split('<span itemprop="height">')
    x=1
    for part in page:
        if x == 3:
            strip = part.split("</span>")
            strip[0] = str(strip[0]).replace("</span>","")
            height = strip[0]
            height = str(height).replace("-","'")+'"'
            x += 1
        else:
            x += 1
    
    df = pd.DataFrame({"height":[height], 
                      "name":[name],
                         },index = [y])
    return df
    

print(heights("Zion Williamson",1))
```

```python
y = 1

df = pd.DataFrame({"height":[],
                   "name":[],
                         },index = [])
for player in players:
    df = df.append(heights(player, y))
    y += 1
```
